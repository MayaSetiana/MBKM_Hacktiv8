# WEIGHTED MEAN

# CORRELATION COEFFICIENT



- r > 0 : relasi positive
- r < 0 : relasi negative
- r = 1 : paling maksimum linear correlation
- r = -1

- r = 1 atau r = -1 atau mendekati merupakan tanda kedua variabel mengalami multicollinearity



```python
var_x = np.var(x, ddof=1)
var_y =  np.var(y, ddof=1)
cov_xy = np.cov(x,y)
cov_xy = cov_xy[0,1]
std_x = np.std(x, ddof=1)
std_y = np.std(y, ddof=1)
r = cov_xy / (std_x * std_y)
cov_xy
r
```
