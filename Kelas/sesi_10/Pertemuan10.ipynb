# Uniform Distribution

- point : distribusi probabilitas dimana setiap nilai pada suatu rentang data memiliki probabilitas yang sama untuk terjadi karena persebaran datanya merata dalam rentang tersebut.

Uniform distribution adalah suatu distribusi probabilitas yang mengasumsikan bahwa nilai-nilai dalam suatu rentang tertentu memiliki probabilitas yang sama untuk terjadi. Distribusi ini juga sering disebut sebagai distribusi persebaran seragam karena nilai-nilainya tersebar secara merata dalam rentang yang diberikan.

Dalam distribusi uniform, setiap nilai dalam rentang memiliki probabilitas yang sama untuk terjadi, dan probabilitasnya dihitung dengan membagi rentang oleh jumlah nilai yang mungkin. Sebagai contoh, jika kita memiliki rentang nilai antara 1 dan 5, dan ingin menentukan probabilitas terjadinya nilai 3, maka probabilitasnya adalah 1/5, karena terdapat 5 nilai yang mungkin (1, 2, 3, 4, dan 5) dengan probabilitas yang sama.

Distribusi seragam banyak digunakan dalam statistik dan analisis data, terutama dalam pemodelan peristiwa acak yang diharapkan terjadi secara merata dalam suatu rentang tertentu. Contoh penggunaannya termasuk dalam pemodelan harga saham, waktu antara kedatangan pelanggan, atau dalam simulasi proses produksi.


```python
from scipy.stats import uniform
```

```python
data_uniform = uniform.rvs(size=10000, 
            loc=10, scale=10)
data_uniform

# scale = rentang x labelnya 10, ex dari 10 sampe 20
```

Secara lebih spesifik, "uniform.rvs" menghasilkan nilai acak dari distribusi seragam dengan rentang tertentu dan probabilitas yang sama untuk setiap nilai dalam rentang tersebut. Rentang dan parameter lainnya seperti skala dan lokasi dapat ditentukan oleh pengguna. Fungsi ini sangat berguna dalam pengujian hipotesis dan simulasi Monte Carlo, di mana diperlukan pengambilan sampel acak dari distribusi seragam.

Fungsi ini memiliki tiga parameter opsional: loc, scale, dan size.

- loc: parameter ini adalah nilai rata-rata (mean) dari distribusi normal. Jika tidak ditentukan, nilai defaultnya adalah 0.
- scale: parameter ini adalah simpangan baku (standard deviation) dari distribusi normal. Jika tidak ditentukan, nilai defaultnya adalah 1.
- size: parameter ini menentukan jumlah sampel yang akan dihasilkan. Jika tidak ditentukan, nilai defaultnya adalah 1.


```python
sns.distplot(data_uniform)
plt.xlabel('Uniform Distribution')
plt.show()
```

# Hypothesis Testing

Hypothesis testing adalah sebuah metode statistik yang digunakan untuk menguji kebenaran suatu hipotesis atau pernyataan mengenai populasi. Hipotesis ini dapat berupa pernyataan tentang nilai rata-rata, proporsi, atau variansi dalam populasi.

Dalam hypothesis testing, kita memiliki dua hipotesis, yaitu hipotesis nol (null hypothesis) dan hipotesis alternatif. Hipotesis nol (null hypothesis) adalah suatu pernyataan yang kita uji kebenarannya. Hipotesis nol sering kali merujuk pada status quo atau keadaan yang sudah diketahui sebelumnya.

Contoh hipotesis nol dalam penelitian dapat berupa pernyataan bahwa tidak ada perbedaan signifikan antara dua kelompok yang diuji, atau tidak ada hubungan antara dua variabel tertentu.

Dalam proses hypothesis testing, kita akan mencoba membuktikan apakah hipotesis nol dapat diterima atau harus ditolak berdasarkan bukti empiris yang diperoleh dari sampel yang diambil dari populasi. Jika hipotesis nol ditolak, maka hipotesis alternatif yang menyatakan adanya perbedaan atau hubungan yang signifikan antara variabel menjadi lebih diterima.

- kalo p value < 0.05 = mengambil h1

### Perbedaan Z-test dan t-test
Z-test dan t-test adalah jenis uji statistik yang digunakan untuk menguji hipotesis nol tentang parameter populasi berdasarkan sampel. Perbedaan utama antara keduanya adalah z-test digunakan ketika kita mengetahui nilai standar deviasi populasi, sedangkan t-test digunakan ketika kita tidak mengetahui nilai standar deviasi populasi dan hanya memiliki estimasi dari sampel.

Z-test digunakan ketika ukuran sampel relatif besar (biasanya lebih dari 30), sedangkan t-test digunakan ketika ukuran sampel relatif kecil (biasanya kurang dari 30). Dalam t-test, estimasi standar deviasi dari sampel digunakan untuk menggantikan nilai standar deviasi populasi yang tidak diketahui.

Selain itu, distribusi yang digunakan oleh z-test adalah distribusi normal standar (distribusi dengan nilai rata-rata 0 dan standar deviasi 1), sedangkan distribusi yang digunakan oleh t-test adalah distribusi t. Distribusi t mirip dengan distribusi normal, namun memiliki ekor yang lebih lebar dan lebih pendek di tengah. Hal ini disebabkan oleh ketidakpastian yang lebih besar pada estimasi standar deviasi dari sampel ketika ukuran sampel relatif kecil.

Karena t-test mengambil estimasi standar deviasi dari sampel, maka t-test memiliki derajat kebebasan (degree of freedom) yang lebih rendah dibandingkan z-test. Derajat kebebasan adalah ukuran sampel yang digunakan dalam perhitungan t-test dan mempengaruhi tingkat signifikansi dari hasil uji statistik.

Secara singkat, perbedaan antara z-test dan t-test adalah bahwa z-test digunakan ketika nilai standar deviasi populasi diketahui dan ukuran sampel relatif besar, sedangkan t-test digunakan ketika nilai standar deviasi populasi tidak diketahui dan ukuran sampel relatif kecil. Z-test menggunakan distribusi normal standar, sementara t-test menggunakan distribusi t. Derajat kebebasan t-test lebih rendah dibandingkan z-test karena menggunakan estimasi standar deviasi dari sampel.

Fungsi sm.stats.ztest pada library statsmodels digunakan untuk melakukan uji hipotesis terhadap rata-rata sampel dengan menggunakan statistik uji z-score. Output dari fungsi ini adalah nilai z-score dan p-value.

Secara umum, hasil dari sm.stats.ztest terdiri dari:

- Nilai z-score: merupakan nilai statistik uji yang dihitung berdasarkan perbedaan antara rata-rata sampel dengan rata-rata populasi yang diuji. Nilai z-score positif menunjukkan bahwa rata-rata sampel lebih besar dari rata-rata populasi yang diuji, sedangkan nilai z-score negatif menunjukkan sebaliknya.

- P-value: merupakan probabilitas untuk mendapatkan hasil uji yang sama ekstrem atau lebih ekstrem daripada yang diamati jika hipotesis nol benar. Semakin kecil p-value, semakin kuat bukti yang dimiliki untuk menolak hipotesis nol. Sebagai acuan umum, nilai p-value yang kurang dari 0,05 menunjukkan bukti yang cukup kuat untuk menolak hipotesis nol.

```python
bmi_hypo = sm.stats.ztest(males['BMXBMI'].dropna(), females['BMXBMI'].dropna(), alternative='larger')

bmi_hypo

# (z-score, pi value)
```

```python
if bmi_hypo[1] > alpha :
    print ('Accept Null Hypothesis (h0). Mean of Males BMI is not LARGER than Females')
else:
    print ('Reject Null Hypothesis (h0). Mean of Males BMI is LARGER than Females')
```

# Data sama hipotesis beda

```python
bmi_hypo_2s = sm.stats.ztest(males['BMXBMI'].dropna(), females['BMXBMI'].dropna(), alternative='two-sided')

bmi_hypo_2s

# (z-score, pi value)
```

```python
if bmi_hypo_2s[1] > alpha :
    print ('Accept Null Hypothesis (h0)')
else:
    print ('Reject Null Hypothesis (h0)')
```

# Hypothesis Testing Population Proportion

```python
n = 1018
pro_null = 0.52
pro_alt = 0.56
```

```python
pro_hypo = sm.stats.proportions_ztest(n*pro_alt, n, pro_null, alternative='larger')
pro_hypo
```

```python
if pro_hypo[1] > alpha :
    print ('Accept Null Hypothesis (h0)')
else:
    print ('Reject Null Hypothesis (h0)')
```
