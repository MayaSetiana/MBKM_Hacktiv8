# Manipulating Columns

```python
nba = df.copy()
```

```python
nba.info()
```

```python
# Tambah Kolom

nba['difference'] = nba['pts'] - nba['opp_pts']
nba.head()
```

```python
# Penambahan points pangkat 2

nba['pts2'] = nba['pts']**2
nba.head()
```

```python
# perubahan kolom

nba['opp_pts'] = 100
nba['opp_pts'].value_counts()
```

```python
nba.head()
```

```python
nba.columns
```

```python
nba = nba[
    ['gameorder', 'game_id', 'lg_id', '_iscopy', 'year_id', 'date_game',
       'seasongame', 'is_playoffs', 'team_id', 'fran_id', 'pts', 'pts2', 'elo_i',
       'elo_n', 'win_equiv', 'opp_id', 'opp_fran', 'opp_pts', 'opp_elo_i',
       'opp_elo_n', 'game_location', 'game_result', 'forecast', 'notes',
       'difference']
]
```

```python
nba
```

# Rename Columns
